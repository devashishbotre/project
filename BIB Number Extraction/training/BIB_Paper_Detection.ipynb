```python
def preprocess_for_ocr(bib_crop):
    bib_crop_gray = cv2.cvtColor(bib_crop, cv2.COLOR_BGR2GRAY)
    _, bib_crop_bin = cv2.threshold(bib_crop_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    bib_crop_contrast = clahe.apply(bib_crop_bin)
    bib_crop_resized = cv2.resize(bib_crop_contrast, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
    return bib_crop_resized
```

```python
def preprocess_for_ocr(bib_crop):
    bib_crop_gray = cv2.cvtColor(bib_crop, cv2.COLOR_BGR2GRAY)
    _, bib_crop_bin = cv2.threshold(bib_crop_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    bib_crop_contrast = clahe.apply(bib_crop_bin)
    bib_crop_resized = cv2.resize(bib_crop_contrast, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
    return bib_crop_resized
```

```python
import pytesseract
import easyocr
def extract_text_with_tesseract(bib_crop):
    config = '--psm 7 -c tessedit_char_whitelist=0123456789'  # Single line text mode, good for numbers
    text = pytesseract.image_to_string(bib_crop, config=config)
    return text
reader=easyocr.Reader(['en'])
def detect_bib_numbers_with_ocr(image_path):
    # Load the trained YOLO model for inference
    results = bibmodel(image_path)  # Run detection

    # Load image with OpenCV for visualization
    image = cv2.imread(image_path)

    for bbox in results[0].boxes.data:  # Loop through detected objects
        x1, y1, x2, y2, conf, cls = bbox[:6]
        if int(cls) == 0:  # Check if it's the bib_number class
            bib_crop = image[int(y1):int(y2), int(x1):int(x2)]  # Crop the bib region

            # Apply OCR to the cropped region
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            bib_crop_preprocessed = preprocess_for_ocr(bib_crop)

            text = extract_text_with_tesseract(bib_crop_preprocessed)
            print(text)
            result=reader.readtext(bib_crop_preprocessed)
            print(result)
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 1)

    # Display the image with bib number detections
    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
detect_bib_numbers_with_ocr('/content/drive/MyDrive/dataset/images/train/600457b648cbaf018981ff36_jpg.rf.c66f6212ef4c7b42bcb46829088b062d.jpg')
```

```python
import os
def save_bib_cropped(image_path,output_folder):
    os.makedirs(output_folder, exist_ok=True)
      # Run detection

    # Load image with OpenCV for visualization
    image = cv2.imread(image_path)
    results=bibmodel(image)
    i=1
    for bbox in results[0].boxes.data:  # Loop through detected objects
        x1, y1, x2, y2, conf, cls = bbox[:6]
        if conf>=0.4 and int(cls) == 0:  # Check if it's the bib_number class
            bib_crop = image[int(y1):int(y2), int(x1):int(x2)]
            output_path = os.path.join(output_folder, f"{os.path.basename(image_path).split('.')[0]}_bib_{i}.jpg")
            cv2.imwrite(output_path, bib_crop)
        i=i+1

# Example usage
save_bib_cropped("/content/drive/MyDrive/dataset/images/train/600457b648cbaf018981ff36_jpg.rf.c66f6212ef4c7b42bcb46829088b062d.jpg", "/content/dataset/train/images")
```

```python
import shutil

def delete_folder(folder_path):
  """Deletes a folder and its contents.

  Args:
    folder_path: The path to the folder you want to delete.
  """
  try:
    shutil.rmtree(folder_path)
    print(f"Folder '{folder_path}' deleted successfully.")
  except OSError as e:
    print(f"Error deleting folder '{folder_path}': {e}")

# Example usage:
folder_to_delete = "/content/dataset"  # Replace with the actual folder path
delete_folder(folder_to_delete)
```

```python
def create_label_file(cropped_images_folder, output_label_file):
    label_entries = []
    for filename in sorted(os.listdir(cropped_images_folder)):
        if filename.endswith(".jpg"):
            image_path = os.path.join(cropped_images_folder, filename)
            print(image_path)
            bib_number = input(f"Enter bib number for {filename}: ")
            label_entries.append(f"{image_path} {bib_number}")

    with open(output_label_file, "w") as f:
        f.write("\n".join(label_entries))

# Generate labels
create_label_file('/content/dataset/train/images', '/content/dataset/train/labels.txt')
```
